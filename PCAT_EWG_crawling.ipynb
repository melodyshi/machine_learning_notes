```python
import time
from selenium import webdriver
import pickle
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
#options.add_argument("--headless") # Runs Chrome in headless mode.
options.add_argument('--no-sandbox') # Bypass OS security model
options.add_argument('--disable-gpu')  # applicable to windows os only
options.add_argument('start-maximized') #
options.add_argument('disable-infobars')
options.add_argument("--disable-extensions")
driver = webdriver.Chrome(chrome_options=options)
```

```python
def get_comp_name(text):
    name = ''
    count = 0
    start = False
    stop = False
    for char in text:
        if char=='(':
            return name
        if start:
            name += char
        if char == ' ':
            start = True   
```

```python
def get_products(company, driver):
    comp_list = []
    url = 'https://www.ewg.org/skindeep/search.php?search_group=companies&query=%3D%3D&&showmore=companies&atatime=50#.W2jbd9JKiUl'
    driver.get(url)
    while True:
        try:
            td_list = driver.find_elements_by_xpath("//td[@align='left']")
            for td in td_list:
                comp_list.append(get_comp_name(td.text).strip())
            atags = driver.find_element_by_xpath("//div[@id='click_next_number' and @class='light']").find_elements_by_tag_name('a')
            found = False
            for a in atags:
                print(a.text)
                if a.text == 'Next>':
                    found = True
                    count += 1
                    print('page'+str(count)+' clicked.')
                    a.click()
                    time.sleep(2)
            if not found
                break

        except:
            print('fail to click page'+str(count))
            pass
        


```
