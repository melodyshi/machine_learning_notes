```python
def get_chem_dict(df):
    df_dict = {}
    chem_dict = {}
    chem_list = df['Chemical'].unique()
    for chem in chem_list:
        chem_dict[df[df.Chemical == chem]['CAS'].unique()[0]] = chem
    df_dict['tri_id'] = df['TRI ID'].unique()[0]
    df_dict['chemicals'] = chem_dict
    return df_dict
```
